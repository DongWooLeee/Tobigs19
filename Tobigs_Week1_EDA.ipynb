```python
# 계속 크로스탭 생성할 것이므로 함수화
import matplotlib.gridspec as gridspec

# 나중에도 쓸 것이므로 함수로 저장
def write_text(ax, total_size):
    for patch in ax.patches:
        ax.text(x = patch.get_x() + patch.get_width()/2,
               y = patch.get_height() + total_size*0.001,
               s = f'{(patch.get_height()/total_size)*100: 1.1f}%', #소수점 1자리까지
               ha = 'center')
        
def get_crosstab(df, feature):
    cross = pd.crosstab(df[feature], df['Attrition'], normalize = 'index')*100
    cross.reset_index(inplace = True)
    return cross
#---------------- 시각화 그래프 함수 생성 ---------------

def draw_two_ratio_plot(fea_list, df, row_num, col_num, size = (15, 18)):
    import matplotlib.gridspec as gridspec

    plt.figure(figsize = size)
    plt.subplots_adjust(wspace = 0.4, hspace = 0.3) #각 ax별 간격 
    grid = gridspec.GridSpec(row_num, col_num)

    #------------------------ 왼쪽 축
    for ind, col in enumerate(fea_list):
        # 크로스탭 생성
        ct = get_crosstab(df, col)
    
        # 특정 ax에 그래프를 그림
        ax = plt.subplot(grid[ind])
        sns.countplot(x = col, data = df, ax = ax, color = 'skyblue', order = ct[col].values)
    
        #text는 앞에서 만든 함수 활용
        write_text(ax, len(df))
    
        # 각 ax에 제목 달아주기
        ax.set_title(f'Distribution of {col}')
    
    
    #----------------------- 오른쪽 축
        # 이중 축 생성후 pointplot을 그림 
        ax2 = ax.twinx()
        ax2 = sns.pointplot(x = ct[col], y = ct[1], color = 'black', order = ct[col].values) # ct[1] 은 target =1 의 비율 
    
        # y축 범위 설정 - 좀 넉넉하게 잡아줌 
        ax2.set_ylim(ct[1].min()-5, ct[1].max()*1.1)
    
        # y축 이름 달아주기
        ax2.set_ylabel('Attrition Ratio')


    plt.rc('font', size = 10)
    plt.show()

draw_two_ratio_plot(cat_cols, data,4,2)

```

그래프 해석)

각 bar 위에 표현된 퍼센테이지는, 단순히 전체 사원 대비 count를 센 것이다. 이렇게 단순히 변수별 특징만을 살펴보는 것이 아니라, 변수와 target간의 관계를 살펴보는 것이 중요하다.

Overtime 유무에 따라 단순히 count로 세었을 때, 그 숫자만을 살펴보면,

퇴사한 사람 자체는 Overtime 유무에 상관 없이 그 수 자체는 비슷하다.
그래서 이 방식으로 본다면, 비율하게 월등하게 큰 고유값들(여기서는 OverTime = No가 될 것이다)에서 절대적인 개수가 많기 때문에, target= 1 인 비율도 더 크게 나올 것이다.

하지만 특정 고유값만 따로 뽑아서 target=1인 비율만 따로 구해보면, 다른 고유값이 target=1을 갖고 있는 비율이 더 클 수도 있다.
즉, 각 고유값 count를 전체로보고, 각 고유값내에서의 attrition=1의 비율을 살펴볼 필요가 있는 것이다. 

이를 보기 위해, pointplot으로서 각 고유값 cnt 중 attrition의 비율을 살펴보았다(그 정확한 수치는 오른쪽 축에서 읽을 수 있다)

이와 같은 통계적 함정에 빠지는 것을 막기 위해,위와 같이 pandas의 crosstab을 활용하여 사용자 지정 시각화 함수를 만들어 사용하였다. 

보면 모든 범주형 변수에서 Attrition=1의 비율이 고유값별로 다르게 나타나고 있으며, 이는 target에 대한 설명력이 있는 변수라고 해석할 수 있다...

예상대로, 전체적인 downwardtrend를 그리는 것을 알 수 있습니다. 즉, 만족도가 높을수록 퇴사비율은 전반적으로 하락하는 경향성을 보입니다.

하지만 본 파생 변수에는 한계가 존재하는데, 이렇게 생성한 파생변수는 상기했듯 기존의 다른 '만족도 관련' 변수들의 가중치가 동일하게 반영되었다는 문제점이 있습니다.

따라서, 다른 변수들의 영향 또한 고려를 해보아야한다는 결론을 내릴 수 있습니다.

또 만족도가 가장 높은 16에서는 퇴사 비율이 매우 높아집니다.
